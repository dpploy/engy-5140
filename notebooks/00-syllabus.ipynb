**Catalog description:** <br>
History of nuclear waste disposal; engineering design of disposal systems. Present status of waste and the character and quantities of future wastes. Review of disposal concepts on a generic basis. The national plan for waste disposal.<br>    
**Prerequisites:** none.<br>
**Co-requisites:** none.<br>
**Course designation:** elective. <br>

**Course outcomes:**
 1. Explain fundamentals of waste processing.
 2. Explain waste management practices.
 3. Understand the basic physicochemical processes in waste processing.
 4. Explain stabilization and solidification.
 5. Describe basic thermal methods.
 6. Understand site remediation and risk analysis.
 7. Explain nuclear waste forms.
 8. Understand nuclear fuel reprocessing and radioactive waste management.
